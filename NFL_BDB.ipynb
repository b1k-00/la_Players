There are only 6 Louisiana colleges that have active players in the NFL this Year. We need to put this in a list

```python
la_colleges = ['Louisiana State', 'Louisiana Tech', 'Louisiana-Lafayette', 'Southern U.', 'Tulane', 'Grambling']
```

Show

```python
la_players = players_df[players_df['collegeName'].isin(la_colleges)]
print(la_players.to_string(index=False))
```

```python
# List the collegeName in aplhabetical order

la_players.sort_values(by=['collegeName'], inplace=True)
print(la_players.to_string(index=False))
```

```python
la_players.sort_values(by=['collegeName'], inplace=True)
print((la_players.to_string(index=False)))
```

# Number of Players by Position

```python
la_players['position'].value_counts().plot(kind = 'bar', color = 'darkblue')
plt.xlabel('Position')
plt.ylabel('Number of Players')
_ = plt.title('Louisiana College Players by Position')
```

```python
# @title LineGraph of LA Players & collegeName

college_counts = la_players.groupby('collegeName').size()
ax = college_counts.plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
ax.set_xlabel('Number of Players')
ax.set_ylabel('College Name')
ax.set_title('Active NFL Players From Louisiana Colleges')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```

# Pie w/ HBCU combined

```python
# Merge Southern U & Grambling in pie chart

combined_count = la_players[la_players['collegeName'].isin(['Southern U.', 'Grambling'])].shape[0]
other_counts = la_players[~la_players['collegeName'].isin(['Southern U.', 'Grambling'])].groupby('collegeName').size()

# Create a pie chart
plt.pie([combined_count] + list(other_counts), labels=['HBCUs'] + list(other_counts.index), autopct="%1.1f%%")
plt.title("Number of Players from Each College")
plt.show()
```

# Pie w/ Legend

```python
# Combine Southern U & Grambling in pie chart above

combined_count = la_players[la_players['collegeName'].isin(['Southern U.', 'Grambling'])].shape[0]
other_counts = la_players[~la_players['collegeName'].isin(['Southern U.', 'Grambling'])].groupby('collegeName').size()

# Create a pie chart
plt.pie([combined_count] + list(other_counts), labels=['HBCUs'] + list(other_counts.index), autopct="%1.1f%%")
plt.title("Number of Players from Each College")
plt.legend(['HBCUs'] + list(other_counts.index), loc="upper right")
plt.show()
```
